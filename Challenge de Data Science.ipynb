```python
#Importando bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Lendo o arquivo Json
base_dados = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
base_dados.head()
```

```python
base_dados.columns
```

```python
#Padronizando os dados individualmente
dados_customer = pd.json_normalize(data= base_dados['customer'])
```

```python
#Dados de Customer padronizado
dados_customer
```

```python
#Dados de Phone padronizados
dados_phone = pd.json_normalize(data=base_dados['phone'])
dados_phone
```

```python
#Dados de Internet padronizados
dados_internet = pd.json_normalize(data=base_dados['internet'])
dados_internet
```

```python
#Dados de account padronizados
dados_account = pd.json_normalize(data=base_dados['account'])
dados_account
```

```python
type(base_dados)
```

```python
dados_limpo = base_dados.drop(columns=['customer', 'phone','internet','account'])
dados_limpo
```

```python
#Junção dos dados para um novo DataFrame
base_dados_principal = pd.concat([dados_limpo,dados_customer,dados_phone,dados_internet,dados_account], axis=1)
```
